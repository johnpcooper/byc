```python
def summarize_clones(all_data_df):
    
    date = all_data_df.expt_date[0]
    file = open(f"{date}_clone_report.txt", "w")
    
    for plasmid in all_data_df.plasmid.unique():

        df_gen = all_data_df[all_data_df.plasmid == plasmid]
        print(f"{plasmid}")
        file.write(f"{plasmid}")

        for genotype in df_gen.genotype.unique():
            print(genotype)
            file.write(f"\n{genotype}")
            df_genotype = df_gen[df_gen.genotype == genotype]

            for clone in df_genotype.clone.unique():
                df = df_genotype[df_genotype.clone == clone]
                median_yfp_norm = df.yfp_norm.median()
                median_gfp_norm = df.gfp_norm.median()
                median_rfp_norm = df.rfp_norm.median()
                sub_channel = df.substrate_ctl[df.index.min()].split()[0]
                ctl_channel = df.substrate_ctl[df.index.min()].split()[1]                
                ratio_mean = df.loc[:, f'{sub_channel}_mean'] / df.loc[:, f'{ctl_channel}_mean']
                ratio_norm = df.loc[:, f'{sub_channel}_norm'] / df.loc[:, f'{ctl_channel}_norm']
                print(f"""
    Clone {clone} 
    YFP norm = {median_yfp_norm}
    RFP norm = {median_rfp_norm}
    GFP norm = {median_gfp_norm}
    {sub_channel}_mean:{ctl_channel}_mean = {ratio_mean.median()}
    {sub_channel}_norm:{ctl_channel}_norm = {ratio_norm.median()}
    """)
                
                file.write(f"""
    Clone {clone} 
    YFP norm = {median_yfp_norm}
    RFP norm = {median_rfp_norm}
    GFP norm = {median_gfp_norm}
    {sub_channel}_mean:{ctl_channel}_mean = {ratio_mean.median()}
    {sub_channel}_norm:{ctl_channel}_norm = {ratio_norm.median()}
""")
```

```python
master_df = pd.read_csv(ssa.select_file())# Choose the master index .csv for you flow experiment
dfs_list = ssa.set_flow_cyto_dfs_list(master_df)
all_data_df = pd.concat(dfs_list, sort=False, ignore_index=True)
```

```python
all_data_df.loc[:, 'yfp_rfp'] = all_data_df['Comp-YFP-A'] / all_data_df['Comp-DsRed-A']
all_data_df.loc[:, 'rfp_yfp'] = all_data_df['Comp-DsRed-A'] / all_data_df['Comp-YFP-A']
```

```python
plasmid = 'pJC072'
filetype='png'
df = all_data_df.loc[all_data_df.plasmid == plasmid]

df.loc[:, 'yfp_rfp_norm'] = df.yfp_rfp/df[df.genotype == 'pdr5d'].yfp_rfp.median()
df.loc[:, 'rfp_yfp_norm'] = df.rfp_yfp/df[df.genotype == 'pdr5d'].rfp_yfp.median()

y_var = 'rfp_yfp_norm'
ylabel = f'{df.substrate_fluor[df.index[0]]}/{df.ref_fluor[df.index[0]]}'
y_lim = (-0.2, 5)

x_var = 'genotype'
xlabel = 'Genotype'

order = ['pdr5d', 'rpn4d']
sample_n = 180


fig = plt.figure(figsize=(1.8, 2.3))
fig.set_dpi(300)

ax = sns.boxplot(x='genotype', y=y_var, hue='clone', data=df,
                 dodge=True, showfliers=False, color='black',
                 linewidth=1, order=order)

facecolors = ['green', 'green', 'green', 'purple', 'purple', 'purple']


    
for i,box in enumerate(ax.artists):
    box.set_facecolor(facecolors[i])
    box.set_alpha(.5)
    box.set_edgecolor('black')
    
for line in ax.lines:
    line.set_color('black')
    line.set_alpha(1)

#ax = sns.swarmplot(x='genotype', y=y_var, hue='clone', data=df.sample(sample_n),
#                   dodge=True, scolor='white', edgecolor='black',
#                   linewidth=1, order=order)


ax.set_ylim(y_lim)
ax.set_ylabel(ylabel, fontsize=11)
ax.set_xlabel(xlabel, fontsize=11)

sns.despine()
ax.legend_.remove()

#fig.savefig(f'{plasmid}_{y_var}.{filetype}')
```
