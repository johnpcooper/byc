```python
def set_fp(prompt):
    
    """ Return the path to a file of interest. Call with the prompt you would like to 
    display in the dialog box."""
    
    # create the dialog box and set the fn
    root = tk.Tk()
    fp = dia.askopenfilename(parent=root, title=prompt)
    root.destroy() # very important to destroy the root object, otherwise python 
    # just keeps running in a dialog box
    
    return fp # return the path to the file you just selected
```

```python
def set_image_stack(prompt):
    
    """ Return a tf.imread() object referring to the image stack"""
    
    fp = set_fp(prompt)
    print("Path to the stack .tif selected:\n%s" % fp)
    image_stack = tf.imread(fp)
    # let the user see the tif stack they just instantiated
    print("Displaying 0th image in this stack:")
    io.imshow(image_stack[0])
    
    return image_stack
```

### How to apply threshold

```python
from skimage.filters import threshold_minimum
from skimage.filters import threshold_otsu

def apply_min_threshold(image, show_plot):

    thresh = threshold_minimum(image)
    binary = image > thresh
    
    if show_plot == True:    

        fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
        ax = axes.ravel()

        ax[0].imshow(image, cmap=plt.cm.gray)
        ax[0].set_title('Original image')

        ax[1].imshow(binary, cmap=plt.cm.gray)
        ax[1].set_title('Minimum Threshold')

        for a in ax:
            a.axis('off')

        plt.show()
    
    return binary
    
def apply_otsu(image, show_plot):

    thresh = threshold_otsu(image)
    binary = image > thresh
    
    if show_plot == True:    

        fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
        ax = axes.ravel()

        ax[0].imshow(image, cmap=plt.cm.gray)
        ax[0].set_title('Original image')

        ax[1].imshow(binary, cmap=plt.cm.gray)
        ax[1].set_title('Otsu Threshold')

        for a in ax:
            a.axis('off')

        plt.show()
        
    else: pass
    
    return binary

    
```

### Watershed Segmentation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi

from skimage.morphology import watershed
from skimage.feature import peak_local_max

def watershed_segementation(image, plot_results, watershed_line):
    
    # Now we want to separate the two objects in image
    # Generate the markers as local maxima of the distance to the background
    distance = ndi.distance_transform_edt(image)
    local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((3, 3)),
                                labels=image)
    
    markers = ndi.label(local_maxi)[0]
    labels = watershed(-distance, markers, mask=image, watershed_line=watershed_line)
    
    if plot_results == True:
    
        fig, axes = plt.subplots(ncols=3, figsize=(9, 3), sharex=True, sharey=True)
        ax = axes.ravel()

        ax[0].imshow(image, cmap=plt.cm.gray, interpolation='nearest')
        ax[0].set_title('Overlapping objects')
        ax[1].imshow(-distance, cmap=plt.cm.gray, interpolation='nearest')
        ax[1].set_title('Distances')
        ax[2].imshow(labels, cmap=plt.cm.nipy_spectral, interpolation='nearest')
        ax[2].set_title('Separated objects')

        for a in ax:
            a.set_axis_off()

        fig.tight_layout()
        plt.show()
        
    return labels
```

```python
binary = watershed_segementation(otsu_thresholded_stack[0], plot_results=True, watershed_line=True)
```

```python
watershed_otsu_thresholded_stack = []

for i in range(0, len(otsu_thresholded_stack)):
    
    image = otsu_thresholded_stack[i]
    binary_image = watershed_segementation(image,
                                           plot_results=False,
                                           watershed_line=True)
    
    watershed_otsu_thresholded_stack.append(binary_image)
```

```python
watershed_otsu_binary_stack = io.concatenate_images(watershed_otsu_thresholded_stack)
io.imsave("watershed_otsu_binary_stack.tif", watershed_otsu_binary_stack)
```
